Pasar la estructura del modelo a Pytorch

```python
# En PyTorch no se define InputLayer explícitamente.
# Tener en cuenta que el tensor de entrada debe tener esta forma:
# (batch_size, 1, 32, 32, 256)

# Corroborar el batch_size
# Corroborar valores de kernel_size, stride, padding
# Corroborar cuál es la función de activación (Se usó ReLu)
# Corroborar el modo de upsampling (bilinear)

class DepthMap(nn.Module):
    def __init__(self):
        super(DepthMap, self).__init__()
        # Encoder 3D
        self.conv3d_2 = nn.Conv3d(1, 48, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU(inplace=True)

        self.conv3d_3 = nn.Conv3d(48, 48, kernel_size=3, stride=1, padding=1)
        self.activation_2 = nn.ReLU()

        self.avgpool3d_1 = nn.AvgPool3d(kernel_size=(1, 1, 256), stride=(1, 1, 256))

        self.avgpool3d_2 = nn.AvgPool3d(kernel_size=(2, 2, 4), stride=(2, 2, 4))

        self.conv3d_5 = nn.Conv3d(48, 96, kernel_size=3, stride=1, padding=1)
        self.activation_3 = nn.ReLU()

        self.conv3d_6 = nn.Conv3d(96, 96, kernel_size=3, stride=1, padding=1)
        self.activation_4 = nn.ReLU()

        self.avgpool3d_3 = nn.AvgPool3d(kernel_size=(1, 1, 4), stride=(1, 1, 4))

        self.avgpool3d_4 = nn.AvgPool3d(kernel_size=(2, 2, 4), stride=(2, 2, 4))

        self.conv3d_8 = nn.Conv3d(96, 192, kernel_size=3, stride=1, padding=1)
        self.activation_5 = nn.ReLU()

        self.conv3d_9 = nn.Conv3d(192, 192, kernel_size=3, stride=1, padding=1)
        self.activation_6 = nn.ReLU()

        self.avgpool3d_5 = nn.AvgPool3d(kernel_size=(2, 2, 4), stride=(2, 2, 4))

        self.avgpool3d_6 = nn.AvgPool3d(kernel_size=(2, 2, 4), stride=(2, 2, 4))

        self.conv3d_11 = nn.Conv3d(192, 384, kernel_size=3, stride=1, padding=1)
        self.activation_7 = nn.ReLU()

        self.conv3d_12 = nn.Conv3d(384, 384, kernel_size=3, stride=1, padding=1)
        self.activation_8 = nn.ReLU()

        self.avgpool3d_7 = nn.AvgPool3d(kernel_size=(2, 2, 16), stride=(2, 2, 16))

        self.avgpool3d_8 = nn.AvgPool3d(kernel_size=(4, 2, 2), stride=(4, 2, 2))

        self.conv3d_14 = nn.Conv3d(384, 768, kernel_size=3, stride=1, padding=1)
        self.activation_9 = nn.ReLU()

        self.conv3d_15 = nn.Conv3d(768, 768, kernel_size=3, stride=1, padding=1)
        self.activation_10 = nn.ReLU()

        self.upsample2d_1 = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)
        self.conv2d_1 = nn.Conv2d(768, 192, kernel_size=3, stride=1, padding=1)

        self.conv2d_2 = nn.Conv2d(576, 384, kernel_size=3, stride=1, padding=1)
        self.activation_11 = nn.ReLU()

        self.conv2d_3 = nn.Conv2d(384, 384, kernel_size=3, stride=1, padding=1)
        self.activation_12 = nn.ReLU()

        self.upsample2d_2 = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)
        self.conv2d_4 = nn.Conv2d(384, 96, kernel_size=3, stride=1, padding=1)

        self.conv2d_5 = nn.Conv2d(288, 192, kernel_size=3, stride=1, padding=1)
        self.activation_13 = nn.ReLU()

        self.conv2d_6 = nn.Conv2d(192, 192, kernel_size=3, stride=1, padding=1)
        self.activation_14 = nn.ReLU()

        self.upsample2d_3 = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)
        self.conv2d_7 = nn.Conv2d(192, 48, kernel_size=3, stride=1, padding=1)

        self.conv2d_8 = nn.Conv2d(144, 96, kernel_size=3, stride=1, padding=1)
        self.activation_15 = nn.ReLU()

        self.conv2d_9 = nn.Conv2d(96, 96, kernel_size=3, stride=1, padding=1)
        self.activation_16 = nn.ReLU()

        self.upsample2d_4 = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)
        self.conv2d_10 = nn.Conv2d(96, 24, kernel_size=3, stride=1, padding=1)

        self.conv2d_11 = nn.Conv2d(72, 48, kernel_size=3, stride=1, padding=1)
        self.activation_17 = nn.ReLU()

        self.conv2d_12 = nn.Conv2d(48, 48, kernel_size=3, stride=1, padding=1)
        self.activation_18 = nn.ReLU()

        self.upsample2d_5 = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=False)
        self.conv2d_13 = nn.Conv2d(48, 12, kernel_size=3, stride=1, padding=1)

        self.conv2d_14 = nn.Conv2d(12, 24, kernel_size=3, stride=1, padding=1)
        self.activation_19 = nn.ReLU()

        self.conv2d_15 = nn.Conv2d(24, 24, kernel_size=3, stride=1, padding=1)
        self.activation_20 = nn.ReLU()

        self.conv2d_16 = nn.Conv2d(24, 1, kernel_size=3, stride=1, padding=1)

        self.flatten = nn.Flatten()
        self.dense_1 = nn.Linear(4096, 4096)
        self.dense_2 = nn.Linear(4096, 4096)

    def forward(self, x):
        x = self.conv3d_2(x)
        x = self.activation_1(x)
        x = self.conv3d_3(x)
        x = self.activation_2(x)

        x1 = self.avgpool3d_2(x)
        x1 = self.conv3d_5(x1)
        x1 = self.activation_3(x1)
        x1 = self.conv3d_6(x1)
        x1 = self.activation_4(x1)

        x2 = self.avgpool3d_1(x)
        x2 = x2.squeeze(-2)  # reshape_5

        x11 = self.avgpool3d_4(x1)
        x11 = self.conv3d_8(x11)
        x11 = self.activation_5(x11)
        x11 = self.conv3d_9(x11)
        x11 = self.activation_6(x11)

        x12 = self.avgpool3d_3(x1)
        x12 = x12.squeeze(3)  # reshape_4

        x112 = self.avgpool3d_5(x11)
        x112 = x112.permute(0, 3, 4, 1, 2)   # reshape_3
        x112 = x112.squeeze(-1)

        x111 = self.avgpool3d_6(x11)
        x111 = self.conv3d_11(x111)
        x111 = self.activation_7(x111)
        x111 = self.conv3d_12(x111)
        x111 = self.activation_8(x111)

        x1112 = self.avgpool3d_7(x111)
        x1112 = x1112.permute(0, 3, 4, 1, 2)  # reshape_2
        x1112 = x1112.squeeze(-1)

        x1111 = self.avgpool3d_8(x111)
        x1111 = self.conv3d_14(x1111)
        x1111 = self.activation_9(x1111)
        x1111 = self.conv3d_15(x1111)
        x1111 = self.activation_10(x1111)
        x1111 = x1111.reshape(x1111.size(0), 2, 2, 768) # reshape_1

        x1111 = x1111.permute(0, 3, 1, 2)
        x1111 = self.upsample2d_1(x1111)
        x1111 = self.conv2d_1(x1111)
        x1111 = x1111.permute(0, 2, 3, 1)

        x1112 = x1112.squeeze(-2)
        x111c = torch.cat((x1111, x1112), dim=-1) # concatenate_1

        x111c = x111c.permute(0, 3, 1, 2)
        x111c = self.conv2d_2(x111c)
        x111c = self.activation_11(x111c)
        x111c = self.conv2d_3(x111c)
        x111c = self.activation_12(x111c)
        x111c = self.upsample2d_2(x111c)
        x111c = self.conv2d_4(x111c)
        x111c = x111c.permute(0, 2, 3, 1)

        x11c = torch.cat((x112, x111c), dim=-1) # concatenate_2

        x11c = x11c.permute(0, 3, 1, 2)
        x11c = self.conv2d_5(x11c)
        x11c = self.activation_13(x11c)
        x11c = self.conv2d_6(x11c)
        x11c = self.activation_14(x11c)
        x11c = self.upsample2d_3(x11c)
        x11c = self.conv2d_7(x11c)
        x11c = x11c.permute(0, 2, 3, 1)

        x1c = torch.cat((x12, x11c), dim=-1) # concatenate_3

        x1c = x1c.permute(0, 3, 1, 2)
        x1c = self.conv2d_8(x1c)
        x1c = self.activation_15(x1c)
        x1c = self.conv2d_9(x1c)
        x1c = self.activation_16(x1c)
        x1c = self.upsample2d_4(x1c)
        x1c = self.conv2d_10(x1c)
        x1c = x1c.permute(0, 2, 3, 1)

        xc = torch.cat((x2, x1c), dim=-1)  # concatenate_4

        xc = xc.permute(0, 3, 1, 2)
        xc = self.conv2d_11(xc)
        xc = self.activation_17(xc)
        xc = self.conv2d_12(xc)
        xc = self.activation_18(xc)
        xc = self.upsample2d_5(xc)
        xc = self.conv2d_13(xc)
        xc = self.conv2d_14(xc)
        xc = self.activation_19(xc)
        xc = self.conv2d_15(xc)
        xc = self.activation_20(xc)
        xc = self.conv2d_16(xc)
        xc = self.flatten(x)
        xc = self.dense_1(x)
        xc = self.dense_2(x)
        xc = xc.view(-1, 64, 64) # reshape_6

        return xc
```
